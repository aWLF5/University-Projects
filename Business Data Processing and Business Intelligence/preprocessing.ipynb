# Load data

```python
import pandas as pd
import glob

# create an empty dictionary to store the dataframes
dfs = {}

# get a list of all csv files in the final_data folder
csv_files = glob.glob('final_data/*.csv')

# loop through each csv file and read it into a separate dataframe
for file in csv_files:
    # read the csv file into a dataframe
    df = pd.read_csv(file)
    # get the name of the csv file without the path and extension
    filename = file.split("/")[-1].split(".")[0]
    # add the dataframe to the dictionary with the filename as the key
    dfs[filename] = df

# pull the dataframes out of the dictionary
for key, value in dfs.items():
    exec(f"{key} = value")
```

```python
# -> break up start_latlng into start_lat and start_lng
# -> break up end_latlng into end_lat and end_lng
segments['start_lat'] = segments['start_latlng'].str.split(',').str[0]
segments['start_lng'] = segments['start_latlng'].str.split(',').str[1]

segments['end_lat'] = segments['end_latlng'].str.split(',').str[0]
segments['end_lng'] = segments['end_latlng'].str.split(',').str[1]

# remove brackets and white spaces from start_lat, start_lng, end_lat, end_lng
segments['start_lat'] = segments['start_lat'].str.replace('[', '')
segments['start_lng'] = segments['start_lng'].str.replace(']', '')
segments['start_lng'] = segments['start_lng'].str.replace(' ', '')

segments['end_lat'] = segments['end_lat'].str.replace('[', '')
segments['end_lng'] = segments['end_lng'].str.replace(']', '')
segments['end_lng'] = segments['end_lng'].str.replace(' ', '')

# drop and rename columns
segments = segments.drop(['Unnamed: 0', 'resource_state', 'start_latlng', 'end_latlng', 'elevation_profile', 'climb_category', 'private', 'hazardous', 'starred'], axis=1)
segments = segments.rename(columns={'id': 'segment_id', 'name': 'segment_name'})
segments

# drop dublicate rows
segments = segments.drop_duplicates(subset=['segment_id'])
```
